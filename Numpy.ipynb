#Multiplicación

```python
n = np.size(Moscu)
y = Moscu
x = fechas
```

```python
a = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2) - np.sum(x)**2)
```

```python
b = np.mean(y)- a*np.mean(x)
```

```python

print(f'La pendiente es: {a}\nLa intersección es: {b}')
```

```python
y = a*x + b
```

```python
y
```

```python
np.linalg.norm(Moscu - y)
```
